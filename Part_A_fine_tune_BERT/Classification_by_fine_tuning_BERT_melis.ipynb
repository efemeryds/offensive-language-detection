```python
import pandas as pd
```

```python
train_data = pd.read_csv('./data/olid-train.csv')
test_data = pd.read_csv('./data/olid-test.csv')
diagnostic_test_data = pd.read_csv('./data/olid-subset-diagnostic-tests.csv')

print(train_data.head())
print(test_data.head())
print(diagnostic_test_data.head())
```

```python
train_data = train_data[['text','labels']]
test_data = test_data[['text','labels']]
diagnostic_test_data = diagnostic_test_data[['text','labels']]
```

```python
print(train_data.head())
print(test_data.head())
print(diagnostic_test_data.head())
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import numpy as np
import logging
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,average_precision_score
from transformers import BertTokenizer,BertModel
from tensorflow.keras.losses import SparseCategoricalCrossentropy
import tensorflow as tf
```

```python
# from https://simpletransformers.ai/docs/usage/ and https://simpletransformers.ai/docs/binary-classification/
model_args = ClassificationArgs()
model_args.num_train_epochs = 15
model_args.learning_rate = 0.002
model_args.overwrite_output_dir= True
model_args.output_dir = '/content/drive/MyDrive/NLP/trained_models'


model = ClassificationModel("bert", 'bert-base-cased', args=model_args)
```

```python
model
```

```python
train_data.isnull().sum()
```

```python
# Train the model

model.train_model(train_data)
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_data, report = classification_report)
```

```python
print(wrong_predictions)
```

```python
predicted_labels = pd.DataFrame({'text':test_data.text, 'predicted': predictions})
predicted_labels.head()
```
